# Extract from CSV

```python
import pandas as pd
data=pd.read_csv("covid19.csv")
data
```

```python
import csv #For Extract from csv file

def Extract(csv_file, columns):
    Extracted_Data = [] #Store the extracted data
    with open(csv_file, 'r') as file:
        txt = csv.reader(file)
        next(txt)  # Skip the header
        for row in txt:
            v = [row[i] for i in columns]
            Extracted_Data.append(v)
    return Extracted_Data

extracted = Extract("covid19.csv",[0, 2, 6, 7, 9, 13, 14])


for x in extracted:
    print(x)
```

```python
df1 = pd.DataFrame(extracted, columns=['ID', 'Age of Subject', 'Time spent on self study', 'Time spent on fitness', 'Time spent on social media', 'Change in your weight', 'Health issue during lockdown'])
df1
```

# Pre-processing (CSV)

```python
#Categorical to numeric - OneHot Encoding (0 --> YES , 1--> No)
df1 = pd.get_dummies(df1, columns=['Health issue during lockdown'])
df1
```

#  Machine Learning

```python
#Clustring to analyzing the relationship between (Time spent on self study,Time spent on social media) 

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X = df1[['Time spent on self study', 'Time spent on social media']]

#Elbow method to determine the best number of clusters 

#Use (WCSS) which is sum of squared distances between each data point 
wcss = []
# k --> from (1-10)
for k in range(1, 11): 
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    wcss.append(k_means.inertia_) 
#(kmeans.inertia_) to calculate the within-cluster sum of squares (WCSS)

# Plot the WCSS values
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
#K-means clustering
k_means = KMeans(n_clusters=3)
k_means.fit(X)

# Get cluster labels for each data point
labels = k_means.labels_
```

```python
# Cluster Distribution Plot
import seaborn as sns

# Count the number of data points in each cluster
cluster_counts = pd.Series(labels).value_counts().sort_index()

# Plot 
plt.figure(figsize=(8, 6))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values)
plt.xlabel('Cluster')
plt.ylabel('Number of Data Points')
plt.title('Cluster Distribution')
plt.show()
```

```python
#Convert [Change in your weight] to numeric to used it in Hierarchical Clustering
df1['Change in your weight']=pd.factorize(df1['Change in your weight'])[0]
#Increased = 0
#Decreased = 1
#Remain Constant = 2
df1
```

```python
# Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage

X = df1[['Age of Subject', 'Time spent on fitness', 'Change in your weight',
        'Health issue during lockdown_NO', 'Health issue during lockdown_YES']]

linked = linkage(X, method='ward')

# Dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked, labels=df1['ID'].values, orientation='top')#labels=df1['ID'].values
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')




plt.tight_layout()  

plt.show()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

# Provide the URL of the website containing the table
url = "https://covid.cdc.gov/covid-data-tracker/#vaccine-effectiveness"
```

```python

# Set up the Chrome driver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode to avoid opening a browser window
service = Service("C:\\Users\\Haneen\\Downloads\\chromedriver_win32")  # Replace with the actual path to chromedriver executable
driver = webdriver.Chrome(service=service, options=chrome_options)

# Navigate to the website
driver.get(url)
```

Using "//table" as the XPath expression is a convenient way to find a table element without specifying its exact location or path in the HTML structure. It will locate the first

element it encounters in the document.

```python
# Wait for the table to load (adjust the sleep time if needed)
driver.implicitly_wait(5)

# Find the table element using its XPath
table_element = driver.find_element(By.XPATH, "//table")

# Extract the table headers
headers = []
header_row = table_element.find_element(By.TAG_NAME, "thead").find_element(By.TAG_NAME, "tr")
for header_cell in header_row.find_elements(By.TAG_NAME, "th"):
    headers.append(header_cell.text.strip())
```

```python
# Extract the table data
table_data = []
for row in table_element.find_elements(By.TAG_NAME, "tr"):
    row_data = [cell.text.strip() for cell in row.find_elements(By.TAG_NAME, "td")]
    if row_data:
        table_data.append(row_data)

# Print the table headers
print(headers)

# Print the table data
for row in table_data:
    print(row)

# Close the driver
driver.quit()
```

```python
import pandas as pd

df2=pd.DataFrame(table_data)
df2.columns=headers
df2
```

```python
# video in arabic
# Step 1 : Importing libararies
import speech_recognition as sr 
import moviepy.editor as mp

# Step 2: Video to Audio conversion
VidClip = mp.VideoFileClip("arabic_video.mp4") 
VidClip.audio.write_audiofile("arabic_audio.wav")

# Step 3: Speech recognition
reco = sr.Recognizer()
audio = sr.AudioFile("arabic_audio.wav")
with audio as source:
  audio_file = reco.record(source)
result = reco.recognize_google(audio_file,language='ar')

# Step 4: Finally exporting the result 
with open('arabic_audio.txt', mode ='w', encoding='utf-8') as file: 
   file.write("Recognized Speech Text:") 
   file.write("\n") 
   file.write(result) 
   print("Text file ready!")
```

```python
from pymongo import MongoClient
import pandas as pd

# Establish MongoDB connection
client = MongoClient('mongodb://localhost:27017/')
db = client['Extract']  # Database name

# Combine data from different resources into a single collection
combined_data = []

# Load data from DataFrame 1
data1 = df1.to_dict('records')
combined_data.extend(data1)

# Load data from DataFrame 2
data2 = df2.to_dict('records')
combined_data.extend(data2)

# Load text from file
with open('arabic_audio.txt', 'r', encoding='utf-8') as file:
    text2 = file.read()
combined_data.append({'text': text2})

# Insert combined data into MongoDB collection
collection = db['combined_collection']  # Collection name

# Delete existing documents in the collection
collection.delete_many({})

# Insert the combined data
collection.insert_many(combined_data)


print('Data sorted and inserted successfully.')
```

```python
# Retrieve all documents from the collection
documents = collection.find()

# empty list to store the retrieved data
data = []

# Iterate over the documents and append them to the data list
for document in documents:
    data.append(document)

# Create a DataFrame from the retrieved data
df_combined = pd.DataFrame(data)


# Display the combined DataFrame
df_combined
```
